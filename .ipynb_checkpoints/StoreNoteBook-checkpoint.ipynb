#### Wrong Data entry

* Two wrong year was entered.
* If this two rows is removed from the table, there will be no sigificant change will be occured to the dataset, because the percentage of data loss is 0.0038%.

* Therefore this two rows is droped from the table

```python
dfc.drop(dfc[(dfc['year']<2011) | (dfc['year']>2014)].index , inplace=True)
dfc[(dfc['year']<2011) | (dfc['year']>2014)]
```

### Analysis

```python
categoryGroup=dfc.groupby(['category']).size().reset_index(name='count')
categoryGroup
```

```python
plt.figure(figsize=(16, 9))
plt.pie(categoryGroup['count'], labels=categoryGroup['category'], autopct='%1.2f%%')
plt.title('Category of sales 2011 - 2014')
plt. legend()
plt.show()
```

```python
countryGroup=dfc.groupby(['country']).size().reset_index(name='count')
countryGroup=countryGroup.sort_values('count',ascending=False)
filterCountry=countryGroup.iloc[:10]
filterCountry=filterCountry.reset_index(drop=True)
#add the rest of the country as others
#filterCountry=filterCountry.append({'country':'Others','count':countryGroup.iloc[20:,1].sum()}, ignore_index=True)
filterCountry
```

```python
plt.figure(figsize=(16, 9))
plt.pie(filterCountry['count'], labels=filterCountry['country'], autopct='%1.2f%%')
plt.title('Country Sales 2011 - 2014')
plt. legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
